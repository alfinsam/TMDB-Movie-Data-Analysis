## Task - 2: Identify the columns that have null values and perform the null value treatment. (Choose the imputation method based on the type of data in the columns of interest)


```python
null_columns = data.columns[data.isnull().any()]
null_column_counts = data[null_columns].isnull().sum()

print("Columns with null values:")
print(null_column_counts)

for column in null_columns:
    if data[column].dtype == 'object':
        data[column].fillna('Unknown', inplace=True)
    else:
        data[column].fillna(data[column].mean(), inplace=True)
null_columns_after_treatment = data.columns[data.isnull().any()]
print("Columns with null values after treatment:")
print(null_columns_after_treatment)
```

## Task - 7: How are popularities of movies related with the movie budgets? Are they correlated or totally uncorrelated with each other? Write the interpretation of your analysis.

```python
correlation_coefficient = data['popularity'].corr(data['budget'])

# Interpret the correlation coefficient
if correlation_coefficient > 0:
    interpretation = "There is a positive correlation between popularity and budget."
elif correlation_coefficient < 0:
    interpretation = "There is a negative correlation between popularity and budget."
else:
    interpretation = "There is no significant correlation between popularity and budget."

# Display the correlation coefficient and interpretation
print(f"Correlation Coefficient: {correlation_coefficient}")
print(interpretation)
```

The correlation coefficient is positive (close to 1), it suggests a positive correlation, meaning that as movie budgets increase, their popularity tends to increase as well.

## Sort the data in descending order based on revenue and filter the top 500 movies. Find the measures of central tendency for the following columns using the filtered data
* Budget
* Revenue
* Runtime
#### Perform outlier analysis for the above three columns using box plots

```python
# Sort the data in descending order based on revenue
sorted_movie_df = data.sort_values(by='revenue', ascending=False)

# Filter the top 500 movies
top_500_movies = sorted_movie_df.head(500)
top_500_movie_names = top_500_movies['title']

# Display the names of the top 500 movies
print("Top 500 Movie Names:")
for movie_name in top_500_movie_names:
    print(movie_name)
    
 
```

```python
# Measures of central tendency
budget_mean = top_500_movies['budget'].mean()
budget_median = top_500_movies['budget'].median()
budget_mode = top_500_movies['budget'].mode().values[0]

revenue_mean = top_500_movies['revenue'].mean()
revenue_median = top_500_movies['revenue'].median()
revenue_mode = top_500_movies['revenue'].mode().values[0]

runtime_mean = top_500_movies['runtime'].mean()
runtime_median = top_500_movies['runtime'].median()
runtime_mode = top_500_movies['runtime'].mode().values[0]

print("Measures of Central Tendency for the Top 500 Movies:")
print(f"BUDGET -  Mean: {budget_mean}    -     Median: {budget_median}    -     Mode: {budget_mode}")
print(f"REVENUE - Mean: {revenue_mean}    -   Median: {revenue_median}    -    Mode: {revenue_mode}")
print(f"RUNTIME - Mean: {runtime_mean}       -       Median: {runtime_median}       -       Mode: {runtime_mode}")

# Outlier analysis using box plots
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.boxplot(top_500_movies['budget'], vert=False)
plt.title('Budget')

plt.subplot(1, 3, 2)
plt.boxplot(top_500_movies['revenue'], vert=False)
plt.title('Revenue')

plt.subplot(1, 3, 3)
plt.boxplot(top_500_movies['runtime'], vert=False)
plt.title('Runtime')

plt.tight_layout()
plt.show()
```
